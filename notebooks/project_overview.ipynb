## Data sources & ingestion

- Playlists tracked and their curator metadata live in `data/playlist_config.json`.
- Artist regional enrichment sits in `data/data/artist_metadata.csv`.
- `scripts/fetch_spotify_data.py` refreshes raw Spotify payloads, normalises them into `data/processed/afrobeats_playlists.json`, and emits `web/data.js` for the dashboards.
- Every ingestion run logs missing artists, curator types, and playlist follower counts so we can audit shifts over time.

## Limitations & next steps

- Spotify currently returns HTTP 403 on every `audio-features` request, so the live dataset excludes tempo/energy metrics; document this limitation explicitly.
- The remaining signals (regional tags, label classification, popularity, follower reach) still cover the research questions on gatekeeping.
- Future experiments: obtain a refreshed user token with the correct scopes for audio features, fold in Apple Music/YouTube comparisons, or extend the artist metadata CSV with manager labels / diaspora granularity.
